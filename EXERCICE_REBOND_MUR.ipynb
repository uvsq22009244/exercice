```python
import tkinter as tk
cpt_balle=1


LARGEUR = 600
HAUTEUR = 400

def deplacement():
    global x, y, z, t
    if z>0 and z<600:

        x=x+20
        z=z+20
        
        canvas.coords(rectangle, x , y , z , t)
        canvas.coords(rectangle1, x2 , y2 , z2 , t2)
        canvas.after(300, deplacement)
        canvas1.after(300, deplacement)
x,y=0,300
z,t=450,420
x4,y4=0,300
z4, t4=450,420

def mouvement():
    rebond()
    canvas.move(balle[0], balle[1], balle[2])
       

def creer_balle():
    """Dessine un rond bleu et retourne son identifiant
     et les valeurs de dÃ©placements dans une liste"""
    x, y = LARGEUR // 2, HAUTEUR // 2
    dx, dy = 3, 5
    rayon = 20
    cercle = canvas.create_oval((x-rayon, y-rayon),
                                (x+rayon, y+rayon),
                                fill="blue")
    return [cercle, dx, dy]


def rebond():
    global balle, cpt_balle, x ,y , t , z, x4 ,y4 , t4, z4
    x0, y0, x1, y1 = canvas.coords(balle[0])
    if x0 <= 0 or x1 >= 600:
        balle[1] = -balle[1]
        cpt_balle+=1  
        
  

        x=x+20
        z=z+20
        
        canvas.coords(rectangle, x , y , z , t)
      
    if y0 <= 0 or y1 >= 300:
        balle[2] = -balle[2]
        cpt_balle+=1
        x4=x4+20
        z4=z4+20
        
        canvas.coords(rectangle1, x4 , y4 , z4 , t4)
        
   
    if cpt_balle<20:
    
        canvas.move(balle[0], balle[1], balle[2])
        canvas.after(20, mouvement)
racine = tk.Tk()

canvas = tk.Canvas(racine, bg="black", width=600, height=400)

rectangle=canvas.create_rectangle(x,y,z,t , fill="blue")
rectangle1=canvas.create_rectangle(x4,y4,z4,t4 , fill="red")

canvas.grid()
balle = creer_balle()
mouvement()


racine.mainloop()
```
